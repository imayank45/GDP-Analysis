GDP Growth on whole dataset

```python
df['Country Name'].unique()
```

```python
final_data = []
for country in df['Country Name'].unique():
  df_pr = df[df['Country Name'] == country]
  gdp_growth_rate = [0]
  data = df_pr.values
  for i in range(1,len(data)):
    curr = data[i][3]
    prev = data[i-1][3]
    gdp_growth_rate.append(round(((curr - prev)/prev)*100,2))
  df_pr = df_pr.assign(GDP = gdp_growth_rate)
  final_data.append(df_pr)
```

```python
final_data[2]
```

```python
df = pd.concat(final_data, axis = 0)
```

```python
import os
```

```python
os.mkdir('GDP Individual Wrt World')
for country in df['Country Name'].unique():

  df_pr = df[df['Country Name'] == country]
  fig = px.line(df_pr, x = 'Year', y = 'Value', title = country+' GDP Analysis Wrt World', range_y = [0,80000000000000])
  pyo.plot(fig, filename = 'GDP Individual Wrt World/'+country+'.html',auto_open = False)
```
